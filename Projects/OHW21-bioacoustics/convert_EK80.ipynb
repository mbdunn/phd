## Convert .raw ek80 to .nc
This step requires version 0.5.0 + of echopype because the framework to convert files has changed.

Also requires Xarray version 0.17.0 at least. Check for combine_attrs error and reinstaller xarray if you get that error.

```python
print(f"echopype version: {ep.__version__}")
print(f"xarray version: {xr.__version__}")
```

### Extract all data from Sailbuoy June 20-29 2018

```python
path = 'D:GLIDER2018/Echosounder/Echosounder'
flist = glob.glob(f"{path}/GLIDER_SB_2018_v5-Phase0-D2018062*.raw")
```

### Convert to netcdf4  

```python
for f_ind in range(len(flist)):
    ed = ep.open_raw(flist[f_ind], sonar_model='ek80');
    ed.to_netcdf(save_path=f"{path}'/nc_files")
```
